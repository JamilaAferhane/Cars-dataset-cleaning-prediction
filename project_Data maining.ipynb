```python
#Diviser la colonne Model en deux
#L'argument n dans split() est utilisé pour spécifier le nombre maximum de séparateurs à utiliser lors de la division d'une chaîne de caractères.
model = cars['Model'].str.split(n=2, expand = True)
```

```python
#Attribuer les colonnes crées aux colonnes Model et Année 
cars['Classe'] = model[0] + " " + model[1]
cars['Année'] = model[2]
# Imprimer l'en-tête et confirmer la création des nouvelles colonnes
cars.head(3000)
```

```python
#Supprimer les colonnes Model et Année
#L'argument inplace dans la fonction drop() est un paramètre booléen qui vous permet de modifier le DataFrame en place,
#sans créer une nouvelle copie du DataFrame.
cars.drop('Model',axis = 1,inplace = True)
cars.drop('Année',axis = 1,inplace = True)
#Imprimer l'en-tête du DataFrame
cars.head()
```

```python
cars['Architecture'] = cars['Architecture'].str.replace("cylindres","")
cars['Architecture'] = cars['Architecture'].str.replace("en ligne","")
cars['Architecture'] = cars['Architecture'].str.replace("en V","")
cars['Architecture'] = cars['Architecture'].astype('float')
cars['Architecture'].mean()
```

```python
# Remove "DH", ".", "VENDUE", "  ", "*", "promo" and "Prix public" from Prices
cars['Price'] = cars['Price'].str.replace("DH","")
cars['Price'] = cars['Price'].str.replace(".","")
cars['Price'] = cars['Price'].str.replace("VENDUE","")
cars['Price'] = cars['Price'].str.replace("  ","")
cars['Price'] = cars['Price'].str.replace('*', '')
cars['Price'] = cars['Price'].str.replace('promo', '')
cars['Price'] = cars['Price'].str.replace('Prix public', '')
# Convert Prices to float
cars['Price'] = cars['Price'].astype('float')
# Print dtypes again
cars.dtypes
```

```python
# Convert Modèle column to datetime
cars['Modèle'] = pd.to_datetime(cars['Modèle'], format='%Y').dt.year
# Print header and datatypes of Modèle column 
print(cars[['Modèle']].head())
```

```python
cars.dtypes
```

```python
# Remove "cv" and "-" from Puissance fiscale
cars['Puissance fiscale'] = cars['Puissance fiscale'].str.replace("cv","")
cars['Puissance fiscale'] = cars['Puissance fiscale'].str.replace("-","")
# Check for missing values in the "puissance fiscale" column
missing_values = cars["Puissance fiscale"].isna()

# Replace non-numeric values with 0.0
cars["Puissance fiscale"].replace(to_replace=r'[^0-9.]', value='0.0', regex=True, inplace=True)
# Clean non-numeric values
cars["Puissance fiscale"] = pd.to_numeric(cars["Puissance fiscale"], errors="coerce")

# Remove leading and trailing white spaces
cars["Puissance fiscale"] = cars["Puissance fiscale"].apply(lambda x: x.strip() if isinstance(x, str) else x)

# Convert Puissance fiscale to float
cars['Puissance fiscale'] = cars['Puissance fiscale'].astype('float')
# Print header and datatypes of Puissance fiscale column 
print(cars[['Puissance fiscale']])
```

```python
cars.dtypes
```

```python
# Remove "cm³" and "." from Cylindrée
cars['Cylindrée'] = cars['Cylindrée'].str.replace("cm³","")
cars['Cylindrée'] = cars['Cylindrée'].str.replace(".","")
# Convert Cylindrée to float
cars['Cylindrée'] = cars['Cylindrée'].astype('float')
# Print header and datatypes of Cylindrée column 
print(cars[['Cylindrée']].head())
```
